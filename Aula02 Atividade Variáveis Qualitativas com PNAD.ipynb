```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python

dados = pd.read_table('pes2013A.txt', sep=' ')
```

```python
dados.head()
```

```python
#Selecionando parte do dataframe - Apenas as variáveis qualitativas
dados.iloc[0:5,0:4]
```

```python
dados.shape
```

```python
n = dados.shape[0]
```

```python
#list(dados)
```

```python
#Variáveis Qualitativas:

dados.Sexo = dados.V0302.astype('float') 

dados.Cor_Raça = dados.V0404.astype('float')

dados.Rede_Ensino = dados.V6002.astype('float')

dados.Nível_Ensino = dados.V4802.astype('float')

dados.Uf = dados.UF.astype('float')

dados.Telefone = dados.V06112.astype('float')

dados.Creche =dados.V0606.astype('float')



```

```python
dados.Sexo.head 
```

```python
print('Frequências relativas por Sexo:')
ut1 = (dados.Sexo.value_counts()/n*100).round(decimals=1).reindex(['4.0', '2.0'])
print(ut1,'\n')

print('Frequências relativas por Cor_Raça:')
ut2 = (dados.Cor_Raça.value_counts()/n*100).round(decimals=1).reindex(['8', '2','4','0','6','9'])
print(ut2,'\n')

print('Frequências relativas por Rede_Ensino:')
ut3 = (dados.Rede_Ensino.value_counts()/n*100).round(decimals=1).reindex(['2.0','4.0'])
print(ut3,'\n')

print('Frequências relativas por Nível_Ensino:')
ut4 = (dados.Nível_Ensino.value_counts()/n*100).round(decimals=1).reindex(['7.0','8.0','10.0','11.0','1.0','6.0','5.0','4.0','2.0','9.0','3.0'])
print(ut4,'\n')

print('Frequências relativas pessoas por estados:')
ut5 = (dados.Uf.value_counts()/n*100).round(decimals=1).reindex(['35', '31','29', '43 ','33','15','26', '23', '41 ','52','13','42', '53', '21',
                                                                 '51','11','32','17','50','25','28','22','27','24','12','16','14'] )
print(ut5,'\n')

print('Frequências relativas Telefone:')
ut6 = (dados.Telefone.value_counts()/n*100).round(decimals=1).reindex(['1.0','3.0'])
print(ut6,'\n')

print('Frequências relativas Creche:')
ut7 = (dados.Creche.value_counts()/n*100).round(decimals=1).reindex(['2.0','4.0'])
print(ut7,'\n')


```

```python
#Gráfico de barras 
plot = ut1.plot(kind='bar',title='Sexo',figsize=(6, 6),color=('blue','green'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Masc/Fem')
```

```python
plot = ut2.plot(kind='bar',title='Cor_Raça',figsize=(6, 6),color=('blue','grey','yellow','purple','black','red'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Cor_Raça')
```

```python
plot = ut3.plot(kind='bar',title='Rede_Ensino',figsize=(6, 6),color=('black','yellow'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Rede_Ensino')
```

```python
plot = ut4.plot(kind='bar',title='Nível_Ensino',figsize=(6, 6),color=('black','yellow'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Nível_Ensino')
```

```python
plot = ut5.plot(kind='bar',title='Pessoas por UF',figsize=(6, 6),color=('black','yellow'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Pessoas por UF')
```

```python
plot = ut6.plot(kind='bar',title='Telefone',figsize=(6, 6),color=('black','yellow'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Pessoas por UF')
```

```python
plot = ut7.plot(kind='bar',title='Creche',figsize=(6, 6),color=('black','yellow'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Creche')
```

```python
#Criar uma coluna 

dados.insert(1,'Regiao',0)
```

```python
dados
```

```python
dados.ix[dados.UF<=17,'Regiao'] = 'Norte'
dados.ix[(dados.UF>=21) & (dados.UF<31),'Regiao'] = 'Nordeste'
dados.ix[(dados.UF>=31) & (dados.UF<41),'Regiao'] = 'Sudeste'
dados.ix[(dados.UF>=41) & (dados.UF<50),'Regiao'] = 'Sul'
dados.ix[(dados.UF>=50),'Regiao'] = 'Centro-Oeste'
```

```python
dados
```

```python
print('Tabela Cruzada Sexo por Regiao:')
ct=pd.crosstab(dados.Sexo,dados.Regiao,margins=True,rownames=['Sexo'], colnames=['Região'])
ct
```

```python
#Tabela Cruzada Sexo- Região
dados.Regiao = dados.Regiao.astype('category')



ct_sem_all =ct.iloc[:2,:5]

ct_sem_all

ct_sem_all = (ct_sem_all/len(dados)*100).round(decimals = 0)
plot = ct_sem_all.plot(kind='bar')
plt.ylabel('Freq.Relativa(em%)')

```
